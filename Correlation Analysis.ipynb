### In this notebook, you learn to simulate a bivariate Gaussian with fixed covariance matrix

Consider two Gaussian variables G1, G2. The covariance matrix of the bivariate Gaussian distribution (G1,G2) is the matrix whose entries are 

[[variance(G1), Covariance(G1,G2)]

[Covariance(G1,G2), variance(G1)]]

```python
covmatrix = [[2, 0],
             [0, 6]]
print(covmatrix)
```

### Consider first the Independent case

```python
covmatrix = [[2, 0],
             [0, 6]]
X, Y = RV(MultivariateNormal(mean = [-1, 1], cov = covmatrix))
(X & Y).sim(10000).plot(alpha = 0.01)
```

Sampling 10000 points according to the distribution shows that no dependency pattern can be identified. Let us compute the empirical values of the covariances for the sample.

```python
(X & Y & X+Y).sim(1000).cov()
```

```python
(X & Y & X+Y).sim(1000).corr()
```

Assume now non zero correlation

```python
covmatrix = [[1, 0.5],
             [0.5, 1]]
X, Y = RV(MultivariateNormal(mean = [-1, 1], cov = covmatrix))
(X & Y).sim(10000).plot(alpha = 0.01)
```

```python
covmatrix = [[1, -0.8],
             [-0.8, 1]]
X, Y = RV(MultivariateNormal(mean = [0, 0], cov = covmatrix))
(X & Y).sim(10000).plot(alpha = 0.01)
```

Analyze the results you observe. Run various simulations by changing the coefficients of the covariance matrices to observe how the distributions are changing.

```python
covmatrix = [[3, -0.1],
             [-0.1, 1]]
X, Y = RV(MultivariateNormal(mean = [0, 0], cov = covmatrix))
(X & Y).sim(10000).plot(alpha = 0.01)
```
